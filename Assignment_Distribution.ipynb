##Question 1 Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

##**ANSWER-->**

To find the probability of rolling a specific number (a 3) exactly 5 times in 30 rolls of a fair die, you use the Binomial Distribution formula.

**Excel Formula: =BINOM.DIST(5, 30, 1/6, FALSE)**

**Parameters :**
* Number_s: 5 (successes)
* Trials: 30 (rolls)
* Probability_s: $1/6$ (probability of rolling a 3 on a single die)
* Cumulative: FALSE (to find the probability of the exact number)*
* **Result: Approximately 0.1921 (or 19.21%).**

##Question 3 A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and 55 if the data follows a normal distribution?  

##**ANSWER-->**

Given a normal distribution with a mean ($\mu$) of 50 and a standard deviation ($\sigma$) of 5:
* Calculation: The range (45 to 55) represents exactly one standard deviation from the mean ($50 - 5$ and $50 + 5$).
* The Empirical Rule: In a normal distribution, approximately 68% of the data points lie within one standard deviation of the mean

##Question 4 What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

##**ANSWER-->**

Standardization is the process of putting different datasets on the same scale.
* Formula: $z = \frac{x - \mu}{\sigma}$
* Transformation: It transforms a dataset so that it has a mean of 0 and a standard deviation of 1.
* Importance: It allows you to compare data points from different distributions (e.g., comparing an SAT score to an ACT score) and helps identify outliers.

##Question 5 What is Kurtosis and their type?

##**ANSWER-->**

Kurtosis measures the "tailedness" or the sharpness of the peak of a probability distribution.

* Mesokurtic: Zero kurtosis; identical to a normal distribution.

* Leptokurtic: Positive kurtosis; sharp peak with heavy, "fat" tails (higher risk of extreme values).

* Platykurtic: Negative kurtosis; flat peak with thin tails (lower risk of extreme values)

##Question 6 Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

##**ANSWER-->**
The uniform distribution is a good model for rolling a fair die because each outcome has an equal probability.
* Explanation: On a 6-sided die, the probability of landing on any specific face (1, 2, 3, 4, 5, or 6) is exactly $1/6$. Since the probability is constant for all possible outcomes, it fits the definition of a discrete uniform distribution.

##Question 7 Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5

##**ANSWER-->**

To find the probability of "at least" 8, you must subtract the cumulative probability of 7 or fewer from 1.

* Excel Formula: =1 - BINOM.DIST(7, 15, 0.5, TRUE)

* Parameters:


* Trials: 15


* Probability: 0.5

* **Result: 0.500 (or 50%). Since 7.5 is the exact middle of 15, the probability of getting 8 or more is equal to getting 7 or less**

##Question 8 How does log transformation help in stabilizing variance and making data more normally distributed?

##**ANSWER-->**

Log transformation is used to handle skewed data.


* Stabilizing Variance: It compresses the scale of the data, reducing the impact of extreme outliers and making the variance more constant across the dataset.


* Normalizing: It pulls in the "long tail" of right-skewed (positively skewed) data, making the distribution look more like a bell curve (normal distribution).
